```python
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
df.info()
```

```python
df["LANGUAGE"].value_counts().head(10).sort_values(ascending=True).plot(kind="barh")
plt.title("Idiomas mais populares")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df["MOST_STREAMED_GAME"].value_counts().head(10).sort_values(ascending=True).plot(kind="line")
plt.title("Jogos mais populares")
plt.xlabel("")
plt.show()
```
